```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./covid.csv', encoding='cp949')
data.head()
```

```python
seoul = data.iloc[:,0:7]
seoul
```

```python
df31 = pd.DataFrame(seoul)

df31=df31[seoul['서울시 기준일'].str.contains('31')]
df31
```

```python
df30 = pd.DataFrame(seoul)

df30=df30[seoul['서울시 기준일'].str.contains('30')]
df30=df30.drop(index=[59,121,180,211,272,333,364,425,486],axis=0)
df30
```

```python
ex= pd.DataFrame(seoul)
ex2 = pd.DataFrame(seoul)
ex3 = pd.DataFrame(seoul)

ex=ex[seoul['서울시 기준일'].str.contains('21.07.28')]
ex2=ex2[seoul['서울시 기준일'].str.contains('20.02.29')]
ex3=ex3[seoul['서울시 기준일'].str.contains('21.02.28')]
```

```python
covid19 = pd.concat([df31,df30,ex,ex2,ex3])
covid19 = covid19.sort_index(ascending=True)
covid19
```

```python
final = pd.DataFrame(covid19)

final = ((final['서울시 확진자'].shift(1) - final['서울시 확진자']))/ final['서울시 확진자']
final
```
